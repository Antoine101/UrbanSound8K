```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample, sample_rate = librosa.load("../dataset/fold1/101415-3-0-2.wav")
s = librosa.feature.melspectrogram(y=sample, sr=22050)
s = librosa.power_to_db(s, ref=np.max)
```

```python
librosa.display.specshow(s)
```

```python
plt.savefig("test", dpi=400, bbox_inches="tight", pad_inches=0)
```

```python
plt.close('all')
```
